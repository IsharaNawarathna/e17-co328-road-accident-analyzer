Here we have created the dataset which we need to train our machine learning time series model.

##Preconfigurations

```python
#import impoartant libraries
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
#getting the folder path to the dataset
dataset_path = "/content/drive/MyDrive/6SP_Machine_Learning_Model/US_Accidents_Dec21_updated.csv"
```

```python
#read the csv file to a dataframe
main_df = pd.read_csv(dataset_path)
```

```python
#lets take a look at the dataframe
main_df.head(10)
```

##Filtering out necessary columns

```python
#creating a new dataset with only start time and start latitude columns
modified_df = main_df.filter(['Start_Time','Start_Lat'], axis=1)
modified_df.head(10)
```

##Dealing with null values

```python
#removing nan values
modified_df = modified_df.dropna(how="any",axis=0)
```

```python
#checking if there is a null value in the dataframe
null_df = modified_df.isnull()
null_df
```

```python
#print if there is a null value
null_df.loc[null_df["Start_Time"]==True]
```

```python
#renaming the column
modified_df.rename(columns = {'Start_Time':'Time'}, inplace = True)
modified_df.head(10)
```

##Working with the date of the dataset and filtering out columns

```python
#convert the Time column to a data time like object
modified_df['Time'] = pd.to_datetime(modified_df['Time'], errors='coerce')
```

```python
#getting the year related to the date
modified_df['Year'] = pd.DatetimeIndex(modified_df['Time']).year
modified_df.head(10)
```

```python
#getting the dataframe which is only have records related to 2017
df_2017 = modified_df.loc[modified_df.Year==2017]
df_2017.head(10)
```

```python
#checking the number of rows in the dataset
len(df_2017)
```

This seems like a good model to work with.

```python
#getting the dataframe into a new variable
df = df_2017
df.tail(10)
```

```python
#dropping the Year column
df = df.drop(["Year"],axis=1)
df.head(10)
```

```python
#getting the Time column for the required format
df['Time'] = df['Time'].dt.strftime('%Y-%m-%d')
df.head(10)
```

##Getting the label of the dataset

```python
#getting the number of accidents related to one date
df_final=df.groupby(by=["Time"],as_index=False).count()
df_final.head(10)
```

```python
#renaming the column
df_final.rename(columns = {'Start_Lat':'NumberOfAccidents'}, inplace = True)
df_final.head(10)
```

```python
#getting the length of the dataframe
len(df_final)
```

```python
#checking if there is a null value in the dataframe
null_df = df_final.isnull()
null_df.loc[null_df["Time"]==True]
```

```python
#creating a csv file with the dataset
df_final.to_csv('time_series_dataset.csv', index=False)
```
